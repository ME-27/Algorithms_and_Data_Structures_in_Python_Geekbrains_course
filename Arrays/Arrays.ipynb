# ДВОИЧНЫЙ (БИНАРНЫЙ) ПОИСК

```python
def bin_search(array, value):
    
    left_border = 0
    right_border = len(array) - 1
    position = len(array) // 2
    
    while array[position] != value and left_border <= right_border:
        
        if value > array[position]:
            left_border = position + 1
        else:
            right_border = position - 1
        
        position = (left_border + right_border) // 2
        
    return -1 if left_border > right_border else position

a = [55, 1, 2, 3, 5, 10, 19, 200]
a.sort()
print(a)

n = int(input('какой элемент найти: '))
print(bin_search(a, n))     
```

# РАЗЛОЖЕНИЕ ПОЛОЖИТЕЛЬНЫХ И ОТРИЦАТЕЛЬНЫХ ЧИСЕЛ ПО РАЗНЫМ МАССИВАМ

хороший вариант - т.к. мы проходимся по списку всего один раз

```python
import random

array = [random.randint(-100, 100) for _ in  range(100)]
print('array: ', array)

negative_numbers_array = []
positive_numbers_array = []

for item in array:
    if item < 0:
        negative_numbers_array.append(item)
    elif item > 0:
        positive_numbers_array.append(item)

print('\nnegative_numbers_array: ', negative_numbers_array)
print('\npositive_numbers_array: ', positive_numbers_array)
```

плохой вариант - т.к. мы проходимся по списку два раза

```python
negative_numbers_array_1 = [item for item in array if item < 0]
positive_numbers_array_1 = [item for item in array if item > 0]

print('\nnegative_numbers_array: ', negative_numbers_array_1)
print('\npositive_numbers_array: ', positive_numbers_array_1)
```

результаты идентичные - поэтому первый вариант лучше

```python
negative_numbers_array == negative_numbers_array_1, positive_numbers_array == positive_numbers_array_1
```

# ВСТАВКА ЭЛЕМЕНТА В ПРОИЗВОЛЬНОЕ МЕСТО МАССИВА

демонстрация работы метода insert

```python
import random

array = [random.randint(-100, 100) for _ in range(10)]
print(array, '\n')

number = 555
position = 2

array.append(None)

i = len(array) - 1
while i > position:
    print(array)
    array[i], array[i - 1] = array[i - 1], array[i]
    i -= 1

array[position] = number
print('\n', array)
```

всё это можно уложить в одну строчку кода

```python
array = [random.randint(-100, 100) for _ in range(10)]
print(array, '\n')

array.insert(position, number)
print(array)
```

как делать на надо - т.к. тратится в два раза больше памяти

```python
array = [random.randint(-100, 100) for _ in range(10)]
print(array, '\n')

array_new = array[:position] + [number] + array[position:]
print(array_new)
```

# ОБМЕН ЗНАЧЕНИЙ ГЛАВНОЙ И ПОБОЧНОЙ ДИАГОНАЛЕЙ КВАДРАТНОЙ МАТРИЦЫ

```python
import random

size = 5

matrix = [[random.randint(1, 10) for _ in range(size)] for _ in range(size)]
for line in matrix:
    for item in line:
        print(f'{item:>4}', end='')  # :>4 - отступ на 4 знака
    print()

# обмен значений
for i in range(size):
    for j in range(size):
        if i == j:
            
            spam = matrix[i][j]
            matrix[i][j] = matrix[i][size - 1 - j]
            matrix[i][size - 1 - j] = spam

# вывод изменённой матрицы, где диагонали поменялись местами
print('\n')
for line in matrix:
    for item in line:
        print(f'{item:>4}', end='')  # :>4 - отступ на 4 знака
    print()
```

В диапазоне натуральных чисел от 2 до 99 определить, сколько из них кратны каждому из чисел в диапазоне от 2 до 9. Примечание: 8 разных ответов.

```python
# формируем словарь ответов
answer = {}
for i in range(2, 10):
    answer.update({i: 0})

# подсчитываем
for i in range(2, 100):
    for j in range(2, 10):     
        if i % j == 0:
            answer[j] += 1
            
# выводим ответы
for i in answer.keys():
    print(f'числу {i} кратны {answer[i]} натуральных чисел от 2 до 99')
```

__Задание 6__

В одномерном массиве найти сумму элементов, находящихся между минимальным и максимальным элементами. Сами минимальный и максимальный элементы в сумму не включать.

```python
import random


array = [random.randint(-10, 10) for _ in range(10)]
print(array, '\n')

min_element = array[0]
max_element = array[0]
min_element_index = 0
max_element_index = 0

for index, element in enumerate(array):
    if min_element > element:
        min_element = element
        min_element_index = index
    if max_element < element:
        max_element = element
        max_element_index = index

print('минимальный элемент: ', min_element)
print('максимальный элемент: ', max_element)
print('индекс минимального элемента: ', min_element_index)
print('индекс максимального элемента: ', max_element_index)

if min_element_index > max_element_index:
    min_element_index, max_element_index = max_element_index, min_element_index

sum_of_required_elements = 0
for element in array[min_element_index + 1:max_element_index]:
    sum_of_required_elements += element

print('\nсумма элементов, находящихся между минимальным и максимальным элементами: ', sum_of_required_elements)
```

```python
import random


array = [random.randint(1, 10) for _ in range(10)]
print(array, '\n')

first_element = None
second_element = None
for element in array:
    if first_element is None:
        first_element = element
    elif first_element > element:
        second_element = first_element
        first_element = element
    elif second_element is None or second_element > element:
        second_element = element

print('первый наименьший элемент: ', first_element)
print('второй наименьший элемент: ', second_element)
```

__Задание 8__

Матрица 5x4 заполняется вводом с клавиатуры, кроме последних элементов строк. Программа должна вычислять сумму введенных элементов каждой строки и записывать ее в последнюю ячейку строки. В конце следует вывести полученную матрицу.

```python
[i for i in range(1, 4)]
```

```python
import random

vertical_length = 4
horizontal_length = 5
matrix = []

for _ in range(1, vertical_length + 1):
    matrix_line = []
    matrix_line_sum = 0
    for _ in range(1, horizontal_length):
        matrix_element = random.randint(1, 10)
        matrix_line.append(matrix_element)
        matrix_line_sum += matrix_element
    matrix_line.append(f'    sum={matrix_line_sum}')
    matrix.append(matrix_line)
    
for line in matrix:
    for item in line:
        print(f'{item:>4}', end='')  # :>4 - отступ на 4 знака
    print()
```

__Задание 9__

Найти максимальный элемент среди минимальных элементов столбцов матрицы.

```python
import random


# генерация матрицы
size = 3
matrix = []
for _ in range(size):
    matrix_line = []
    for _ in range(size):
        matrix_element = random.randint(1, 10)
        matrix_line.append(matrix_element)
    matrix.append(matrix_line)

# вывод матрицы
for line in matrix:
    for item in line:
        print(f'{item:>4}', end='')
    print()


# рассчёт минимальных элементов столбцов матрицы
min_elements_in_columns = [matrix[0][i] for i in range(size)]
for line in matrix:
    for element_position, element in enumerate(line):
        if min_elements_in_columns[element_position] > element:
            min_elements_in_columns[element_position] = element

# вывод минимальных элементов столбцов матрицы
print('\nминимальные элементы столбцов матрицы:')
for item in min_elements_in_columns:
    print(f'{item:>4}', end='')

# рассчёт максимального элемента среди минимальных элементов столбцов матрицы
max_element_in_min_elements_in_columns = min_elements_in_columns[0]
for element in min_elements_in_columns:
    if max_element_in_min_elements_in_columns < element:
        max_element_in_min_elements_in_columns = element

print('\n\nмаксимальный элемент среди минимальных элементов столбцов матрицы: ', max_element_in_min_elements_in_columns)
```
