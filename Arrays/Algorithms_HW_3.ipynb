__Задание 1__

В диапазоне натуральных чисел от 2 до 99 определить, сколько из них кратны каждому из чисел в диапазоне от 2 до 9. Примечание: 8 разных ответов.

```python
# формируем словарь ответов
answer = {}
for i in range(2, 10):
    answer.update({i: 0})

# подсчитываем
for i in range(2, 100):
    for j in range(2, 10):     
        if i % j == 0:
            answer[j] += 1
            
# выводим ответы
for i in answer.keys():
    print(f'числу {i} кратны {answer[i]} натуральных чисел от 2 до 99')
```

__Задание 2__

Во втором массиве сохранить индексы четных элементов первого массива. Например, если дан массив со значениями 8, 3, 15, 6, 4, 2, второй массив надо заполнить значениями 0, 3, 4, 5 (помните, что индексация начинается с нуля), т. к. именно в этих позициях первого массива стоят четные числа.

```python
import random

array_1 = [random.randint(-100, 100) for _ in range(10)]
print('первый массив: ', array_1)

array_2 = []
for index, element in enumerate(array_1):
    if element != 0 and element % 2 == 0:
        array_2.append(index)
    
print('второй массив: ', array_2)
```

__Задание 3__

В массиве случайных целых чисел поменять местами минимальный и максимальный элементы.

первый вариант

```python
import random

array = [random.randint(-100, 100) for _ in range(10)]
print('изначальный массив\n: ', array, '\n')

min_element = array[0]
min_element_index = 0
max_element = array[0]
max_element_index = 0
for index, element in enumerate(array):
    if min_element > element:
        min_element = element
        min_element_index = index
    if max_element < element:
        max_element = element
        max_element_index = index
        
array[min_element_index] = max_element
array[max_element_index] = min_element

print('минимальный элемент: ', min_element)
print('максимальный элемент: ', max_element, '\n')

print('массив, где минимальный и максимальный элементы поменялись местами\n: ', array)
```

__Задание 4__

Определить, какое число в массиве встречается чаще всего.

```python
import random


array = [random.randint(0, 5) for _ in range(10)]
print(array, '\n')

the_most_frequent_number = array[0]
max_number_of_entries = 0
for element in array:
    number_of_element_entries = 0
    for i in array:
        if element == i:
            number_of_element_entries += 1
    if max_number_of_entries < number_of_element_entries:
        max_number_of_entries = number_of_element_entries
        the_most_frequent_number = element

print('ответ: ', the_most_frequent_number)
```

__Задание 5__

В массиве найти максимальный отрицательный элемент. Вывести на экран его значение и позицию в массиве.

```python
import random


array = [random.randint(-5, 5) for _ in range(10)]
print(array, '\n')

max_negative_number = None
max_negative_number_index = None

for index, element in enumerate(array):
    if element < 0:
        if max_negative_number == None:
            max_negative_number = element
            max_negative_number_index = index
        elif max_negative_number < element:
            max_negative_number = element
            max_negative_number_index = index
print(f'максимальный отрицательный элемент: {max_negative_number}\nего позиция: {max_negative_number_index}')
```

__Задание 6__

В одномерном массиве найти сумму элементов, находящихся между минимальным и максимальным элементами. Сами минимальный и максимальный элементы в сумму не включать.

```python
import random


array = [random.randint(-10, 10) for _ in range(10)]
print(array, '\n')

min_element = array[0]
max_element = array[0]
min_element_index = 0
max_element_index = 0

for index, element in enumerate(array):
    if min_element > element:
        min_element = element
        min_element_index = index
    if max_element < element:
        max_element = element
        max_element_index = index

print('минимальный элемент: ', min_element)
print('максимальный элемент: ', max_element)
print('индекс минимального элемента: ', min_element_index)
print('индекс максимального элемента: ', max_element_index)

if min_element_index > max_element_index:
    min_element_index, max_element_index = max_element_index, min_element_index

sum_of_required_elements = 0
for element in array[min_element_index + 1:max_element_index]:
    sum_of_required_elements += element

print('\nсумма элементов, находящихся между минимальным и максимальным элементами: ', sum_of_required_elements)
```

__Задание 7__

В одномерном массиве целых чисел определить два наименьших элемента. Они могут быть как равны между собой (оба минимальны), так и различаться.

```python
import random


array = [random.randint(1, 10) for _ in range(10)]
print(array, '\n')

first_element = None
second_element = None
for element in array:
    if first_element is None:
        first_element = element
    elif first_element > element:
        second_element = first_element
        first_element = element
    elif second_element is None or second_element > element:
        second_element = element

print('первый наименьший элемент: ', first_element)
print('второй наименьший элемент: ', second_element)
```

__Задание 8__

Матрица 5x4 заполняется вводом с клавиатуры, кроме последних элементов строк. Программа должна вычислять сумму введенных элементов каждой строки и записывать ее в последнюю ячейку строки. В конце следует вывести полученную матрицу.

```python
import random

vertical_length = 4
horizontal_length = 5
matrix = []

for _ in range(1, vertical_length + 1):
    matrix_line = []
    matrix_line_sum = 0
    for _ in range(1, horizontal_length):
        matrix_element = random.randint(1, 10)  # заполнил randint вместо input, чтобы быстрее
        matrix_line.append(matrix_element)
        matrix_line_sum += matrix_element
    matrix_line.append(f'    sum={matrix_line_sum}')
    matrix.append(matrix_line)
    
for line in matrix:
    for item in line:
        print(f'{item:>4}', end='')  # :>4 - отступ на 4 знака
    print()
```

__Задание 9__

Найти максимальный элемент среди минимальных элементов столбцов матрицы.

```python
import random


# генерация матрицы
size = 3
matrix = []
for _ in range(size):
    matrix_line = []
    for _ in range(size):
        matrix_element = random.randint(1, 10)
        matrix_line.append(matrix_element)
    matrix.append(matrix_line)

# вывод матрицы
for line in matrix:
    for item in line:
        print(f'{item:>4}', end='')
    print()


# рассчёт минимальных элементов столбцов матрицы
min_elements_in_columns = matrix[0]
for line in matrix[1:]:
    for element_position, element in enumerate(line):
        if min_elements_in_columns[element_position] > element:
            min_elements_in_columns[element_position] = element

# вывод минимальных элементов столбцов матрицы
print('\nминимальные элементы столбцов матрицы:')
for item in min_elements_in_columns:
    print(f'{item:>4}', end='')

# рассчёт максимального элемента среди минимальных элементов столбцов матрицы
max_element_in_min_elements_in_columns = min_elements_in_columns[0]
for element in min_elements_in_columns:
    if max_element_in_min_elements_in_columns < element:
        max_element_in_min_elements_in_columns = element

print('\n\nмаксимальный элемент среди минимальных элементов столбцов матрицы: ', max_element_in_min_elements_in_columns)
```
