```python
from random import randint, uniform


random_value_data_type = input('укажите тип данных диапазона значений \n' +
                               'a - случайное целое число, \n' +
                               'b - случайное вещественное число, \n' +
                               'c - случайный символ \n: ')

if random_value_data_type == 'a':
    lowest_value = int(input('укажите нижнюю границу диапазона значений: '))
    highest_value = int(input('укажите верхнюю границу диапазона значений: '))

    print('вывод случайного целого числа: ', randint(lowest_value, highest_value))

elif random_value_data_type == 'b':
    lowest_value = float(input('укажите нижнюю границу диапазона значений: '))
    highest_value = float(input('укажите верхнюю границу диапазона значений: '))

    print('вывод случайного вещественного числа: ', uniform(lowest_value, highest_value))

elif random_value_data_type == 'c':
    lowest_value = ord(input('укажите нижнюю границу диапазона значений: '))
    highest_value = ord(input('укажите верхнюю границу диапазона значений: '))
    
    print('вывод случайного символа: ', chr(randint(lowest_value, highest_value)))
```

```python
side_1 = float(input('введите длину первой стороны треугольника: '))
side_2 = float(input('введите длину второй стороны треугольника: '))
side_3 = float(input('введите длину третьей стороны треугольника: '))

if max(side_1, side_2, side_3) < side_1 + side_2 + side_3 - max(side_1, side_2, side_3):
    if side_1 != side_2 and side_1 != side_3 and side_2 != side_3:
        print('треугольник разносторонний')
    elif side_1 == side_2 and side_1 == side_3:
        print('треугольник равносторонний')
    else:
        print('треугольник равнобедренный')
else:
    print('такой треугольник не существует')
```

```python
number_1 = float(input('введите первое число: '))
number_2 = float(input('введите второе число: '))
number_3 = float(input('введите третье число: '))

if number_1 != max(number_1, number_2, number_3) and number_1 != min(number_1, number_2, number_3):
    print(f'средним числом является {number_1}')

elif number_2 != max(number_1, number_2, number_3) and number_2 != min(number_1, number_2, number_3):
    print(f'средним числом является {number_2}')

else:
    print(f'средним числом является {number_3}')
```
