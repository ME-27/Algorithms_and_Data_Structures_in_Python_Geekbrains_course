## Задание 1

Пользователь вводит данные о количестве предприятий, их наименования и прибыль за четыре квартала для каждого предприятия. Программа должна определить среднюю прибыль (за год для всех предприятий) и отдельно вывести наименования предприятий, чья прибыль выше среднего и ниже среднего.

```python
from collections import defaultdict


companies = defaultdict(int)
for _ in range(int(input('введите количество компаний: '))):
    company_name = input('\nвведите название компании: ')
    for i in range(4):
        companies[company_name] += int(input(f'введите прибыль компании "{company_name}" за {i + 1} квартал: '))

mean_profit = sum(companies.values()) / len(companies.keys())
print('\nсредняя прибыль: ', mean_profit)

companies_with_profit_under_average = []
companies_with_profit_above_average = []

for key, value in companies.items():
    if value <= mean_profit:
        companies_with_profit_under_average.append(key)
    else:
        companies_with_profit_above_average.append(key)
        
print('\nкомпании с прибылью ниже среднего: ')
for i in companies_with_profit_under_average:
    print(i)

print('\nкомпании с прибылью выше среднего: ')
for i in companies_with_profit_above_average:
    print(i)
```

## Задание 2

Написать программу сложения и умножения двух шестнадцатеричных чисел. При этом каждое число представляется как массив, элементы которого — цифры числа. Например, пользователь ввёл A2 и C4F. Нужно сохранить их как [‘A’, ‘2’] и [‘C’, ‘4’, ‘F’] соответственно. Сумма чисел из примера: [‘C’, ‘F’, ‘1’], произведение - [‘7’, ‘C’, ‘9’, ‘F’, ‘E’]. Примечание: Если воспользоваться функциями hex() и/или int() для преобразования систем счисления, задача решается в несколько строк. Для прокачки алгоритмического мышления такой вариант не подходит. Поэтому использование встроенных функций для перевода из одной системы счисления в другую в данной задаче под запретом. Вспомните начальную школу и попробуйте написать сложение и умножение в столбик.

```python
from collections import deque


# стандартный ввод
# hexadecimal_number_1 = deque(input('введите первое шестнадцатиричное число: '))
# hexadecimal_number_2 = deque(input('введите второе шестнадцатиричное число: '))

# для быстроты присвоил значения программно
hexadecimal_number_1 = deque('a2')
hexadecimal_number_2 = deque('c4f')

if len(hexadecimal_number_1) < len(hexadecimal_number_2):
    hexadecimal_number_1, hexadecimal_number_2 = hexadecimal_number_2, hexadecimal_number_1
    
def hexadecimal_numbers_addition(hexadecimal_number_1_, hexadecimal_number_2_):
    # функция сложения двух шестнадцатиричных чисел
    
    def hexadecimal_digits_addition(hexadecimal_digit_1_, hexadecimal_digit_2_, hexadecimal_digit_3_):
        # поцифровое сложение шестнадцатиричных цифр с возвратом остатка, а также 1/0 в следующий цикл
        
        scale = deque('0123456789abcdef')

        sum_decimal = scale.index(hexadecimal_digit_1_) +\
                      scale.index(hexadecimal_digit_2_) +\
                      scale.index(hexadecimal_digit_3_)
        
        if sum_decimal > 15:
            return '1', scale[sum_decimal - 16]
        else:
            return '0', scale[sum_decimal]
    
    remainder = '0'
    hexadecimal_numbers_sum = deque()
    while True:
        try:
            hexadecimal_digit_1 = hexadecimal_number_1_.pop()
        except:
            hexadecimal_digit_1 = '0'
        try:
            hexadecimal_digit_2 = hexadecimal_number_2_.pop()
        except:
            hexadecimal_digit_2 = '0'

        if hexadecimal_digit_1 == '0' and hexadecimal_digit_2 == '0' and remainder == '0':
            break
        else:
            remainder, hexadecimal_digit_to_answer =\
                hexadecimal_digits_addition(hexadecimal_digit_1, hexadecimal_digit_2, remainder)
            hexadecimal_numbers_sum.appendleft(hexadecimal_digit_to_answer)

    return hexadecimal_numbers_sum

print('\nсумма введённых шестнадцатиричных чисел: ')
print(hexadecimal_numbers_addition(hexadecimal_number_1.copy(), hexadecimal_number_2.copy()))


# при умножении использовал подход отличный от использованного выше:
# создал две функции - по переводу из шестнадцатиричного в десятиричный формат и обратно
def hexadecimal_numbers_multiplication(hexadecimal_number_1_, hexadecimal_number_2_):
    # функция умножения двух шестнадцатиричных чисел
    
    def from_hexadecimal_to_decimal(hexadecimal_number):
        # функция перевода числа из шестнадцатиричного в десятиричный формат
        decimal_number = 0
        for index, value in enumerate([deque('0123456789abcdef').index(i) for i in hexadecimal_number][::-1]):
            decimal_number += value * (16 ** index)
        return decimal_number

    def from_decimal_to_hexadecimal(decimal_number):
        # функция перевода числа из десятиричного в шестнадцатиричный формат
        hexadecimal_number = deque()
        while decimal_number != 0:
            hexadecimal_number.appendleft(deque('0123456789abcdef')[decimal_number % 16])
            decimal_number //= 16
        return hexadecimal_number
    
    return from_decimal_to_hexadecimal(from_hexadecimal_to_decimal(hexadecimal_number_1_) *\
                                       from_hexadecimal_to_decimal(hexadecimal_number_2_))

print('\nпроизведение введённых шестнадцатиричных чисел: ')
print(hexadecimal_numbers_multiplication(hexadecimal_number_1, hexadecimal_number_2))
```
