# КАК ФУНКЦИОНИРУЕТ ПАМЯТЬ В LIST

```python
allocated = 0

for newsize in range(100):
    if allocated < newsize:
        new_allocated = (newsize >> 3) + (3 if newsize < 9 else 6)
        allocated = newsize + new_allocated
    print(newsize, allocated)
```

# СКОЛЬКО ПАМЯТИ ЗАНИМАЮТ РАЗЛИЧНЫЕ ОБЪЕКТЫ

```python
import sys

print(sys.version, sys.platform)
```

```python
a = 5
b = 125.54
c = 'Hello world!'
```

```python
print(sys.getsizeof(a))
print(sys.getsizeof(b))
print(sys.getsizeof(c))
```

размеры указаны в байтах

```python
my_list = [i for i in range(10)]
print(sys.getsizeof(my_list))
my_list = [i for i in range(100)]
print(sys.getsizeof(my_list))
my_list = [i for i in range(1000)]
print(sys.getsizeof(my_list))
```

```python
my_list = [i for i in range(10)]

def show_size(x, level=0):
    print('\t' * level, f'type= {x.__class__}, size= {sys.getsizeof(x)}, object= {x}')
    
    if hasattr(x, '__iter__'):
        if hasattr(x, 'items'):
            for xx in x.items():
                show_size(xx, level + 1)
        elif not isinstance(x, str):
            for xx in x:
                show_size(xx, level + 1)
                
show_size(a)
show_size(b)
show_size(c)
show_size(my_list)
```

```python
t = tuple(my_list)
show_size(t)
```

кортеж занимает в памяти 120 байт, в то время как список занимает 184

```python
s = set(my_list)
show_size(s)
```

множество занимает аж 728 байт!!!

поиск элементов внутри множества осуществляется за константное время

в списке и кортеже поиск будет зависить от количества элементов в структуре

```python
d = {str(i): i for i in range(10)}
d
```

```python
show_size(d)
```

# СТРУКТУРА ОБЪЕКТА В ПАМЯТИ

```python
import sys
import ctypes
import struct


a = 5
b = 125.54
c = 'Hello world!'

print(id(a))  # адрес в памяти по которому лежит объект целого типа 5
print(sys.getsizeof(a))  # 28 байт

print(ctypes.string_at(id(a), sys.getsizeof(a)))  # 28 байт
print(struct.unpack('LLLLLLl', ctypes.string_at(id(a), sys.getsizeof(a))))
```

```python
a = 59
b = 125.54
c = 'Hello world!'

print(id(a))  # адрес в памяти по которому лежит объект целого типа 5
print(sys.getsizeof(a))  # 28 байт

print(ctypes.string_at(id(a), sys.getsizeof(a)))  # 28 байт
print(struct.unpack('LLLLLLl', ctypes.string_at(id(a), sys.getsizeof(a))))
```

счётчик ссылок (первое значение) уменьшился, а значение (последняя цифра) стала 59

```python
a = 59
x = y = a
b = 125.54
c = 'Hello world!'

print(id(a))  # адрес в памяти по которому лежит объект целого типа 5
print(sys.getsizeof(a))  # 28 байт

print(ctypes.string_at(id(a), sys.getsizeof(a)))  # 28 байт
print(struct.unpack('LLLLLLl', ctypes.string_at(id(a), sys.getsizeof(a))))
```

счётчик ссылок (первое значение) увеличился на 2 из-за x = y = a
