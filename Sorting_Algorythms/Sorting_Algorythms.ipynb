## СОРТИРОВКА ПУЗЫРЬКОМ

```python
import random

size = 10
array = [i for i in range(size)]
random.shuffle(array)
print(array)
```

```python
n = 1
while n < len(array):
    for i in range(len(array) - n):
        if array[i] > array[i + 1]:
            array[i], array[i + 1] = array[i + 1], array[i]
    n += 1
    print(array)

print('\n')
print(array)
```

## СОРТИРОВКА ВЫБОРОМ

```python
import random

size = 10
array = [i for i in range(size)]
random.shuffle(array)
print(array)
```

```python
def selection_sort(array):
    for i in range(len(array)):
        idx_min = i
        
        for j in range(i + 1, len(array)):  # переменная j идёт по неотсортированной части массива
            if array[j] < array[idx_min]:
                idx_min = j
        
        array[idx_min], array[i] = array[i], array[idx_min]
        print(array)
```

```python
selection_sort(array)

print('\n')
print(array)
```

## СОРТИРОВКА ВСТАВКАМИ

```python
import random

size = 10
array = [i for i in range(size)]
random.shuffle(array)
print(array)
```

```python
def insertion_sort(array):
    
    for i in range(1, len(array)):
        spam = array[i]
        j = i
        
        while array[j - 1] > spam and j > 0:
            array[j] = array[j - 1]
            j -= 1
        
        array[j] = spam
        print(array)
```

```python
insertion_sort(array)

print('\n')
print(array)
```

## СОРТИРОВКА ШЕЛЛА

```python
import random

size = 1000
array = [i for i in range(size)]
random.shuffle(array)
print(array)
```

```python
def shell_sort(array):
    
    assert len(array) < 4000, 'Массив слишком большой, используйте другую сортировку'
    
    def new_increment(array):
        
        inc = [1, 4, 10, 23, 57, 132, 301, 701, 1750]
        
        while len(array) <= inc[-1]:
            inc.pop()
        
        while len(inc) > 0:
            yield inc.pop()
    
    count = 0
    for increment in new_increment(array):
        for i in range(increment, len(array)):
            for j in range(i, increment - 1, -increment):
                if array[j - increment] <= array[j]:
                    break
                array[j], array[j - increment] = array[j - increment], array[j]
                count += 1
#                 print(array)
    print('count: ', count)
```

```python
shell_sort(array)

print('\n')
# print(array)
```

## БЫСТРАЯ СОРТИРОВКА ХОАРА

```python
import random

size = 10
array = [i for i in range(size)]
random.shuffle(array)
print(array)
```

### с использованием дополнительной памяти

```python
def quick_sort(array):
    
    if len(array) <= 1:
        return array
    
    pivot = random.choice(array)
    s_ar = []
    m_ar = []
    l_ar = []
    
    for item in array:
        
        if item < pivot:
            s_ar.append(item)
        elif item > pivot:
            l_ar.append(item)
        elif item == pivot:
            m_ar.append(item)
        else:
            raise Exception('Случилось непредвиденное')
            
    print(s_ar, m_ar, l_ar)
    return quick_sort(s_ar) + m_ar + quick_sort(l_ar)
```

```python
array_new = quick_sort(array)

print('\n')
print(array_new)
```

### без использования дополнительной памяти

```python
import random

size = 10
array = [i for i in range(size)]
random.shuffle(array)
print(array)
```

```python
def quick_sort_no_memory(array, first_, last_):  # first_ - указательна на то, где надо начинать сортировать массив
                                                 # last_ - указательна на то, где надо закончить сортировку
        if first_ >= last_:
            return
        
        print(array)
        
        pivot = array[random.randint(first_, last_)]
        i, j = first_, last_
        
        while i <= j:
            
            while array[i] < pivot:
                i += 1
            
            while array[j] > pivot:
                j -= 1
            
            if i <= j:
                array[i], array[j] = array[j], array[i]
                i, j = i + 1, j - 1
        
        quick_sort_no_memory(array, first_, j)
        quick_sort_no_memory(array, i, last_)
```

```python
quick_sort_no_memory(array, 0, len(array) - 1)

print('\n')
print(array)
```

## РАЗВОРОТ МАССИВА

```python
import random

size = 10
array = [i for i in range(size)]
random.shuffle(array)
print(array)
```

```python
def reverse(array):
    for i in range(len(array) // 2):
        array[i], array[len(array) - i - 1] = array[len(array) - i - 1], array[i]
```

```python
reverse(array)
print(array)
```

## СОРТИРОВКА ПО КЛЮЧУ

```python
from collections import namedtuple


Person = namedtuple('Person', 'name, age')

p_1 = Person('Vasya', 25)
p_2 = Person('Kolya', 22)
p_3 = Person('Olya', 28)

people = [p_1, p_2, p_3]

print(people)
```

```python
people[1]
```

```python
type(people[1])
```

#### простая сортировка

```python
sorted(people)
```

#### 1-й вариант сортировки по ключу

```python
def by_age(person):
    return person.age

sorted(people, key=by_age)
```

#### 2-й вариант сортировки по ключу

```python
sorted(people, key=lambda x: x.age)
```

#### 3-й вариант сортировки по ключу

```python
from operator import attrgetter


sorted(people, key=attrgetter('age'))
```
