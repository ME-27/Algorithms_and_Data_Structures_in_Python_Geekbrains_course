## Задание 1

Отсортируйте по убыванию методом пузырька одномерный целочисленный массив, заданный случайными числами на промежутке [-100; 100). Выведите на экран исходный и отсортированный массивы.
Примечания:

a. алгоритм сортировки должен быть в виде функции, которая принимает на вход массив данных,

b. постарайтесь сделать алгоритм умнее, но помните, что у вас должна остаться сортировка пузырьком. Улучшенные версии сортировки, например, расчёской, шейкерная и другие в зачёт не идут.

#### первый вариант: в виде функции, которая сортирует массив по убыванию, без улучшений в коде сортировки

```python
import random


size = 10
array = [random.randint(-100, 100) for i in range(size)]
random.shuffle(array)
print(f'исходный массив:\n {array}\n')


def bubble_descending_sorting(array):
    n = 1
    while n < len(array):
        for i in range(len(array) - n):
            if array[i] < array[i + 1]:
                array[i], array[i + 1] = array[i + 1], array[i]
                
        print(f'шаг {n} сортировки:\t{array}')
        n += 1
        

bubble_descending_sorting(array)
print(f'\nотсортированный массив:\n {array}')
```

как видно, были пройденты все шаги сортировки, несмотря на то, что в последних шагах ничего не происходило

#### второй вариант: добавлен выход из внешнего цикла при отсутствии действия во внутреннем

```python
import random


size = 10
array = [random.randint(-100, 100) for i in range(size)]
random.shuffle(array)
print(f'исходный массив:\n {array}\n')


def bubble_descending_sorting_improved(array):
    n = 1
    while n < len(array):
        cycle_exit_condition = True
        for i in range(len(array) - n):
            if array[i] < array[i + 1]:
                array[i], array[i + 1] = array[i + 1], array[i]
                cycle_exit_condition = False
        if cycle_exit_condition:
            break
                
        print(f'шаг {n} сортировки:\t{array}')
        n += 1
        

bubble_descending_sorting_improved(array)
print(f'\nотсортированный массив:\n {array}')
```

как видно, внешний цикл был прерван на 7-м шаге сортировки (из 9-ти возможных) в виду отсортированности массива, и программа не работала впустую

## Задание 2

Отсортируйте по возрастанию методом слияния одномерный вещественный массив, заданный случайными числами на промежутке [0; 50). Выведите на экран исходный и отсортированный массивы.

```python
import random


size = 10
array = [random.randint(0, 50) for i in range(size)]
random.shuffle(array)
print(f'исходный массив:\n {array}\n')


def merge_sorting(array):
    n = len(array)
    if n < 2:
        return array
    
    left_part = merge_sorting(array[:(n // 2)])
    right_part = merge_sorting(array[(n // 2):n])
    
    i = j = 0
    result = []
    while i < len(left_part) or j < len(right_part):
        if not i < len(left_part):
            result.append(right_part[j])
            j += 1
        elif not j < len(right_part):
            result.append(left_part[i])
            i += 1
        elif left_part[i] < right_part[j]:
            result.append(left_part[i])
            i += 1
        else:
            result.append(right_part[j])
            j += 1
        
    return result


array = merge_sorting(array)
print(f'\nотсортированный массив:\n {array}')
```

## Задание 3

Массив размером 2m + 1, где m — натуральное число, заполнен случайным образом. Найдите в массиве медиану. Медианой называется элемент ряда, делящий его на две равные части: в одной находятся элементы, которые не меньше медианы, в другой — не больше медианы. Примечание: задачу можно решить без сортировки исходного массива. Но если это слишком сложно, используйте метод сортировки, который не рассматривался на уроках (сортировка слиянием также недопустима).

```python
import random


m = 5
size = 2 * m + 1
array = [random.randint(0, 9) for i in range(size)]
random.shuffle(array)
print(f'исходный массив:\n {array}\n')


def detect_median_in_odd_array(array):
    # функция принимает на вход массив чисел, где количество элементов нечётно, и ищет его медиану.
    # описание работы алгоритма:
    # при входе в функцию изначальный массив делится на 3 части: левую, правую и средний элемент.
    # далее запускается внешний цикл while в котором функционируют ещё 2 внутренних for:
    # первый цикл for пробегается по левой части и меняет местами со средним элементом элементы 
    # если они больше среднего элемента,
    # второй цикл for пробегается по правой части и меняет местами со средним элементом элементы 
    # если они меньше среднего элемента.
    # внешний цикл while крутится до тех пор пока в обеих внутренних циклах не будет никаких перестановок.
    
    left_part = array[:(len(array) // 2)]
    right_part = array[(len(array) // 2):]
    median = right_part.pop(0)
    
    step_counter = 1
    while True:
        
        print(f'\nшаг {step_counter} алгоритма')
        print(f'левая половина: \t{left_part}')
        print(f'правая половина: \t{right_part}')
        print(f'средний элемент: \t\t{median}')
    
        cycle_exit_condition = True
        for i in range(len(left_part)):
            if left_part[i] > median:
                left_part[i], median = median, left_part[i]
                cycle_exit_condition = False
                
        for i in range(len(right_part)):
            if right_part[i] < median:
                right_part[i], median = median, right_part[i]
                cycle_exit_condition = False
        
        if cycle_exit_condition:
            break
        
        step_counter += 1   
    
    print(f'\nмедиана массива: {median}')
    
detect_median_in_odd_array(array)
```
