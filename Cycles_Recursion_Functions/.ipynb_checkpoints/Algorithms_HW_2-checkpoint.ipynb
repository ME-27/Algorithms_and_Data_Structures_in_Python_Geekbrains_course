Договариваемся об идеальном пользователе, который вводит только верные данные, которые требует программа. Проверка ввода не обязательна. Уделите это время в первую очередь построению графического алгоритма, а затем написанию кода по этому алгоритму. Не наоборот.

<img src="Algorithms_HW_2_task_1_diagram.png">

```python
while True:
    operation_symbol = input('введите знак операции (+, -, *, / или 0 для выхода из программы): ')
    if operation_symbol == '0':
        break
    elif operation_symbol != '+' and operation_symbol != '-' and operation_symbol != '*' and operation_symbol != '/':
        print('введён неизвестный знак операции, попробуйте ввести снова')
        continue
        
    number_1 = int(input('введите первое число: '))
    number_2 = int(input('введите второе число: '))
    
    if operation_symbol == '+':
        print(f'{number_1} + {number_2} = {number_1 + number_2}')
    elif operation_symbol == '-':
        print(f'{number_1} - {number_2} = {number_1 - number_2}')
    elif operation_symbol == '*':
        print(f'{number_1} * {number_2} = {number_1 * number_2}')
    elif operation_symbol == '/':
        if number_2 == 0:
            print('делить на 0 нельзя!')
            continue
        print(f'{number_1} / {number_2} = {number_1 / number_2}')
    
```

```python
number = int(input('введите натуральное число: '))

even_number_count = 0
odd_number_count = 0
while number != 0:
    if number % 2 == 1:
        odd_number_count += 1
    else:
        even_number_count += 1
    number //= 10
    
print('количество чётных чисел: ', even_number_count)
print('количество нечётных чисел: ', odd_number_count)
```

```python
number = int(input('введите натуральное число: '))

reverse_number = ''
while number != 0:
    reverse_number += f'{number % 10}'
    number //= 10

print('обратное по порядку число: ', reverse_number)
```

```python
n = int(input('введите число n: ')) - 1

def func(m, n):
    if n != 0:
        return m + func(m / -2, n - 1)
    else:
        return m
    
print(f'сумма элементов ряда чисел: {func(1, n)}')
```

```python
i = 32
while True:
    ii = 1
    line = ''
    while ii <= 10 and i != 127:
        line += f'\t{i}:"{chr(i)}"'
        ii += 1
        i += 1
    print(f'{line}\n')
    if i == 127:
        break
```

```python
from random import randint


number = randint(0, 100)

print('отгадайте число от 0 до 100')
print('у вас 10 попыток\n')

for i in range (1, 11):
    answer = int(input(f'попытка № {i}: '))
    if number == answer:
        print('\nвы угадали!')
        break
    elif number < answer:
        print('введённое число больше\n')
    else:
        print('введённое число меньше\n')

if number != answer: 
    print('\nвы проиграли! правильный ответ: ', number)
```

```python
n = int(input('введите число n: '))

def func(m, n):
    if n != m:
        return m + func(m + 1, n)
    else:
        return m
    
print(f'сумма элементов ряда чисел: {func(1, n)}')
print(f'сумма элементов ряда чисел: {int(n * (n + 1) / 2)}')
```

```python
sequence = int(input('введите последовательность чисел: '))
n = int(input('введите число n: '))
n_count = 0

while sequence != 0:
    if sequence % 10 == n:
        n_count += 1
    sequence //= 10

print(f'цифра {n} встречается в последовательности {n_count} раз(а)')
```
