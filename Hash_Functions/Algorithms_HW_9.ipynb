## Задание 1

Определение количества различных подстрок с использованием хеш-функции. Пусть на вход функции дана строка. Требуется вернуть количество различных подстрок в этой строке. Примечания:
в сумму не включаем пустую строку и строку целиком;
без использования функций для вычисления хэша (hash(), sha1() или любой другой из модуля hashlib задача считается не решённой.

```python
import hashlib


def count_the_number_of_substrings_in_a_string(string):
    # функция принимает на вход строку
    # и обходит при помощи циклов все возможные комбинации подстрок, добавляя при этом
    # в переменную-множество хэши каждой комбинации.
    # таким образом, в переменной-множестве остаются только уникальные хэши и можно посчитать их количество,
    # что и будет количеством уникальных подстрок.
    
    string_indexes = [_ for _ in range(len(string) + 1)]  # формируем список индексов строки
    set_of_substrings = set()  # инициализируем множество для хранения уникальных подстрок
    
    counter = 1  # счётчик для нумерации неуникальных комбинаций (можно и без него)
    for i in string_indexes:
        for ii in string_indexes[i:]:
            if i == ii:  # убираем пустые подстроки
                continue
            elif i == min(string_indexes) and ii == max(string_indexes): # убираем подстроку в виде целой строки
                continue

            print(f"подстрока №{counter}\tиндексы: {i}:{ii}\tподстрока: {string[i:ii]}\t\tхэш: {hashlib.sha1(string[i:ii].encode('utf-8')).hexdigest()}")
            set_of_substrings.add(hashlib.sha1(string[i:ii].encode('utf-8')).hexdigest())  # добавляем полученные хэши в множество
            
            counter += 1

    print(f'\nколичество уникальных подстрок в строке: {len(set_of_substrings)}')
```

проверим работу программы

```python
count_the_number_of_substrings_in_a_string('bom bim bom')
```

как видно, комбинаций было 65, но уникальных подстрок только 53!

## Задание 2

Закодируйте любую строку по алгоритму Хаффмана.

```python
from collections import OrderedDict


def encode_string_huffman_way(string):
    # функция принимает на вход строку в виде string, а возвращает (печатает) её в виде а-ля Хаффман
    
    # разбиваем строку поэлементно
    string = [_ for _ in string]
    print(f'\nстрока поэлементно: {string}')
    
    # получаем словарь частот символов
    string_symbols_frequencies = {}
    for i in list(set(string)):
        string_symbols_frequencies.update({i:string.count(i)})
    
    # убеждаемся, что у нас в строке более чем один уникальный символ
    assert len(string_symbols_frequencies) >= 2, 'в строке не может быть только один уникальный символ!'
    
    # сортируем частоты символов в порядке возрастания 
    string_symbols_frequencies = OrderedDict(sorted(string_symbols_frequencies.items(), key=lambda x: x[1]))
    string_symbols_frequencies = [_ for _ in string_symbols_frequencies.items()]
    print(f'\nчастоты символов в порядке возрастания: {string_symbols_frequencies}')
    
    
    # получаем дерево Хаффмана
    # дерево Хаффмана сохраняем обратно в переменную string_symbols_frequencies!!!
    while True:
        if len(string_symbols_frequencies) >= 2:
            new_element = ([string_symbols_frequencies[0], string_symbols_frequencies[1]], 
                           string_symbols_frequencies[0][1] + string_symbols_frequencies[1][1])
            string_symbols_frequencies = string_symbols_frequencies[2:]
            string_symbols_frequencies.append(new_element)
            string_symbols_frequencies = sorted(string_symbols_frequencies, key=lambda x: x[1])
        else:
            break
    print(f'\nдерево Хаффмана: {string_symbols_frequencies}')
    
    
    # функция для получания кодировочной таблицы из дерева Хаффмана
    def get_symbol_codes_dictionary(huffman_tree, symbols_and_pathes={}, binary_path=''):
        # функция принимает на вход дерево Хаффмана, а возвращает кодировочную таблицу для символов строки
        # функция использует рекурсию для обхода дерева Хаффмана
        
        # если в функцию пришло дерево с двумя листами
        if type(huffman_tree[0][0][0]) == str and type(huffman_tree[0][1][0]) == str:
            symbols_and_pathes.update({huffman_tree[0][0][0]: f'{binary_path}' + '0', 
                                       huffman_tree[0][1][0]: f'{binary_path}' + '1'})
            return symbols_and_pathes
        
        # если в функцию пришло дерево с листом слева и узлом справа
        elif type(huffman_tree[0][0][0]) == str:
            symbols_and_pathes.update({huffman_tree[0][0][0]: f'{binary_path}' + '0'})
            symbols_and_pathes.update(get_symbol_codes_dictionary(huffman_tree[0][1],  # рекурсия!!!
                                                                  symbols_and_pathes, 
                                                                  f'{binary_path}' + '1'))
            return symbols_and_pathes
        
        # если в функцию пришло дерево с листом справа и узлом слева
        elif type(huffman_tree[0][1][0]) == str:
            symbols_and_pathes.update({huffman_tree[0][1][0]: f'{binary_path}' + '1'})
            symbols_and_pathes.update(get_symbol_codes_dictionary(huffman_tree[0][0],  # рекурсия!!! 
                                                                  symbols_and_pathes, 
                                                                  f'{binary_path}' + '0'))
            return symbols_and_pathes
        
        # если в функцию пришло дерево с двумя узлами
        else:
            symbols_and_pathes.update(get_symbol_codes_dictionary(huffman_tree[0][0],  # рекурсия!!! 
                                                                  symbols_and_pathes, 
                                                                  f'{binary_path}' + '0'))
            
            symbols_and_pathes.update(get_symbol_codes_dictionary(huffman_tree[0][1],  # рекурсия!!! 
                                                                  symbols_and_pathes, 
                                                                  f'{binary_path}' + '1'))
            return symbols_and_pathes
    
    
    # получаем кодировочную таблицу
    symbol_codes_dictionary = get_symbol_codes_dictionary(string_symbols_frequencies[0])
    print(f'\nкодировочная таблица: {symbol_codes_dictionary}')
    
    # кодируем строку а-ля Хаффман
    string_encoded_huffman_way = ''
    for symbol in string:
        string_encoded_huffman_way += symbol_codes_dictionary[symbol] + ' '
    
    # кодируем строку в бинарный вид (для сравнения)
    string_encoded_binary = ''
    for symbol in string:
        string_encoded_binary += bin(ord(symbol))[2:] + ' '
    
    
    
    print(f'\nстрока в бинарном виде (для сравнения): {string_encoded_binary}')
    print(f'\n\n\nстрока а-ля Хаффман: {string_encoded_huffman_way}')
```

выведем результат работы написанной функции - строку а-ля Хаффман

```python
encode_string_huffman_way('aaaaaaabbbbbbcccccddddeeeff ')
```

видно, что наименее частотные символы получили более длинные коды
